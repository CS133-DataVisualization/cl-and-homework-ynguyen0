## Problem 1
Convert the following data into tidy format with the headers `country`, `drink` and `servings`. Drop the column `total_litres_of_pure_alcohol`

```python
import pandas as pd

url = "https://raw.githubusercontent.com/fivethirtyeight/data/master/alcohol-consumption/drinks.csv"
df = pd.read_csv(url)

df = df.drop(columns=["total_litres_of_pure_alcohol"])

tidy_df = df.melt(
    id_vars="country",
    var_name="drink",
    value_name="servings"
)

tidy_df.head()
```

## Problem 2
Remove `"_servings"` from the data in the tidy dataframe in Problem 1. Who drinks the most beer, wine and spirits? Create three categorical plots of your choice to illustrate the top 10 countries that drink the most wine, beer or spirits.

```python
import matplotlib.pyplot as plt
import seaborn as sns

tidy_df["drink"] = tidy_df["drink"].str.replace("_servings", "", regex=False)

max_per_drink = tidy_df.loc[tidy_df.groupby("drink")["servings"].idxmax()]
max_per_drink

for drink_type in ["beer", "wine", "spirit"]:
    subset = tidy_df[tidy_df["drink"] == drink_type].nlargest(10, "servings")

    plt.figure(figsize=(10, 6))
    sns.barplot(
        data=subset,
        x="servings",
        y="country",
        order=subset.sort_values("servings", ascending=False)["country"]
    )
    plt.title(f"Top 10 {drink_type.title()} Drinking Countries")
    plt.xlabel("Average servings per person annually")
    plt.ylabel("Country")
    plt.tight_layout()
    plt.show()

```

## Problem 3
Create a wide dataframe using `pivot_table` with the tidy dataframe from Problem 2

```python
wide_df = tidy_df.pivot_table(
    index="country",
    columns="drink",
    values="servings"
).reset_index()

wide_df.head()
```
